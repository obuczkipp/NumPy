# Standard Deviation and Variance

### Demo of std and var

```python
high_var_array = np.array( [100, 200, 500, 4000, 5000])
high_var_array

low_var_array = np.array([2, 4, 6, 7, 9])
low_var_array
```

```python
np.var(high_var_array)
```

```python
np.var(low_var_array)
```

```python
np.std(high_var_array), np.std(low_var_array)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
plt.hist(low_var_array)
```

```python
np.mean(low_var_array)
```

```python
# Átlag, variancia és szórás kiszámítása
mean = np.mean(low_var_array)
variance = np.var(low_var_array)
std_dev = np.std(low_var_array)

# Adatok ábrázolása
plt.plot(low_var_array, label='Data')

# Átlag, variancia és szórás hozzáadása a grafikonhoz
plt.axhline(y=mean, color='r', linestyle='-', label=f'Mean: {mean:.2f}')
plt.axhline(y=mean + std_dev, color='g', linestyle='--', label=f'Std Dev: {std_dev:.2f}')
plt.axhline(y=mean - std_dev, color='g', linestyle='--')
plt.axhline(y=mean + variance, color='b', linestyle=':', label=f'Variance: {variance:.2f}')
plt.axhline(y=mean - variance, color='b', linestyle=':')

# Cím és címkék hozzáadása
plt.title("Low Variance Array with Mean, Variance, and Std Dev")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()

# Grafikon megjelenítése
plt.show()
```

```python
# Create OO method
```

```python
over_50.target.values
```

```python

# set the style
plt.style.use('tableau-colorblind10')

## OO Method from Scratch

fig, ax = plt.subplots(figsize=(10, 6))


# plot data
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'],
                    cmap = "winter") # change the color - season

# customize


ax.set(title= " Heart disase & Cholesterol Levels",
       xlabel= "Age",
       ylabel= "Cholesterol");

# Add Legend * means iteration through each value

ax.legend(*scatter.legend_elements(), title = "Target");

# Add a Horizontal Line to show the mean value

ax.axhline(over_50["chol"].mean(),
          linestyle= "--", color = "red", label = "Mean Cholesterol Level "
          )
ax.legend();
```

```python
over_50.head()
```

```python
plt.style.available
```

```python
# Subplot of chol, age, tha..


fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))
# two axises ax0, ax1 

# Add to data to ax0

scatter = ax0.scatter(x = over_50['age'],
                      y= over_50['chol'],
                      c= over_50['target'])

# Customize ax0

ax0.set(title= "Heart",
        xlabel= "Age",
        ylabel= "Cholesterol")

# add title

ax0.legend(*scatter.legend_elements(), title = "Target")

# Add Meanline

ax0.axhline(y=over_50['chol'].mean(),
linestyle = "--",
color = "red",
label = "Mean Value"
);

# Add data to ax1

scatter = ax1.scatter( x= over_50['age'],
                       y= over_50['thalach'],
                      c= over_50['target'])

# customize ax1

ax1.set( title= "Title Dieasese and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate");

# Add legend to ax1

ax1.legend(*scatter.legend_elements(), title = "Target")


# add max line

ax1.axhline(y= over_50['thalach'].max(),
            linestyle = "--",
            color = "red",
            label = "Max Heart Rate"
           );
        
# Add main title to figure

fig.suptitle("Heart Analysis", fontsize=22, fontweight = "bold") # Main title
```
